# Hypothesis

# H0 :  mean of Repurchase Price = mean of Net Asset Value (NAV)
# H1 : mean of Repurchase Price != mean of Net Asset Value (NAV)
Hypothesis Second
# H0 :  mean of Sale Price = mean of Net Asset Value (NAV)
# H1 : mean of Sale Price != mean of Net Asset Value (NAV)



```python
import scipy.stats as stats
from statsmodels.stats import weightstats as stests
#Using 2 Sample T-Test 
#checking relation between Repurchase Price and Sale Price

_,pvalue=stats.ttest_ind(a=dataclean_1['Repurchase Price'], b=dataclean_1['Net Asset Value'], equal_var=True)
print("2 Sample T test p Value is", pvalue)
if( pvalue < 0.05) :
  print("Rejecting Null Hypothesis")
else:
  print("Accepting Null Hypothesis")  

#Using 2 Sample T-Test 
#checking for linear relation between Sale Price and NAV
_,pvalue=stats.ttest_ind(a=dataclean_1['Sale Price'], b=dataclean_1['Net Asset Value'], equal_var=True)
print("2 Sample T test p Value is", pvalue)
if( pvalue < 0.05) :
  print("Rejecting Null Hypothesis")
else:
  print("Accepting Null Hypothesis")  
#Z test  
ztest ,pval1 = stests.ztest(dataclean_1['Repurchase Price'], x2=dataclean_1['Net Asset Value'], value=0,alternative='two-sided')
print(float(pval1))
if pval1<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
ztest ,pval1 = stests.ztest(dataclean_1['Sale Price'], x2=dataclean_1['Net Asset Value'], value=0,alternative='two-sided')
print(float(pval1))
if pval1<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis") 
```
